```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="dataset/data.yaml",   # YAML inside dataset/
    epochs=200,
    imgsz=1024,
    batch=4,       # reduce if OOM on RTX4050
    device=0,
    patience=30,
    hsv_h=0.02,
    hsv_s=0.5,
    hsv_v=0.5,
    mosaic=1.0,
)
```
